```python
result = 0
for x in input:
  a, b =  x.split('-')

  for k in range(int(a), int(b)+1):
    n = str(k)
    i = len(n)
    if i % 2 == 0:
      mid = i // 2
      if n[:mid] == n[mid:]:
        result += int(k)
result
```

```python
result = 0
for x in input:
  a, b = map(int, x.split("-"))

  for m in range(a, b+1):
    s = str(m)
    n = len(s)
    for k in range(1, n):
        if n % k == 0:
            sub = s[:k]
            if sub * (n // k) == s:
              result += m
              break

result
```
